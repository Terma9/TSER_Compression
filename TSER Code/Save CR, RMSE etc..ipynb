Steps:

```python
# Save for each wavelet, the rmses for each dataset for each decomp level
# Add dropout-value of 0 -> Explain for thesis, 0 no compression no transform

wavelets = ['bior3.5', 'haar', 'db4']


for split in splits: 
    for wavelet in wavelets:

        for ds_name in ds_names:

            dataset_array = np.load('/home/sim/Desktop/TS Extrinsic Regression/data/Dataset-Arrays/' + ds_name + f'_{split}.npy')
            max_level = pywt.dwt_max_level(dataset_params[ds_name]['block_size'], wavelet)

            for level in range(1, max_level + 1):
                rmses = np.empty_like(dropout_values)

                for i, dropout in enumerate(dropout_values):
                    if dropout == 0:
                        rmses[i] = 0 # No Compression, so rmse of 0
                    else:
                        rmses[i] = compute_avg_rmse_of_dataset(dataset_array, compress_dataset(dataset_array.copy(), ds_name, andDecompress= True, compression_type="dwt", compression_param=dropout, wavelet=wavelet, level=level))



                np.save('/home/sim/Desktop/TS Extrinsic Regression/data/Wavelet-Levels/' + f'wv-lvl:{level}-{ds_name}_{wavelet}_{split}', rmses)
```

```python
#Save the rmse for each wavelet and its best lvl for each dataset
wavelets = ['bior3.5', 'haar', 'db4']

for split in splits: 
    for wavelet in wavelets:

        for ds_name in ds_names:

            dataset_array = np.load('/home/sim/Desktop/TS Extrinsic Regression/data/Dataset-Arrays/' + ds_name + f'_{split}.npy')

            #Depending on former results, believe max level
            max_level = pywt.dwt_max_level(dataset_params[ds_name]['block_size'], wavelet)  

            rmses = np.empty_like(dropout_values)

            for i, dropout in enumerate(dropout_values):
                if dropout == 0:
                    rmses[i] = 0 
                else:
                    rmses[i] = compute_avg_rmse_of_dataset(dataset_array, compress_dataset(dataset_array.copy(), ds_name, andDecompress= True, compression_type="dwt", compression_param=dropout, wavelet=wavelet, level=max_level))



            np.save('/home/sim/Desktop/TS Extrinsic Regression/data/Winner-Wavelet/' + f'wv-win-{ds_name}_{wavelet}_{split}', rmses)



```

```python
#Plot the Quantisation Levels
#-> Check with the going into negative and so on!
# I can only run, after having winner wavelet and levels defined!
# -> I also need to change the big code to acitvate quantisation


for split in splits: 
    for tq in tqs:

        for ds_name in ds_names:

            dataset_array = np.load('/home/sim/Desktop/TS Extrinsic Regression/data/Dataset-Arrays/' + ds_name + f'_{split}.npy')


            # I decide how many ranges we test
            for precision in range(8):

                comp_ratios = np.empty_like(dropout_values)
                rmses = np.empty_like(dropout_values)


                for i, dropout in enumerate(dropout_values):
                    if dropout == 0:
                        rmses[i] = 0 
                        comp_ratios[i] = 1
                    else:
                        rmses[i] = compute_avg_rmse_of_dataset(dataset_array, compress_dataset(dataset_array.copy(), ds_name, andDecompress=True, compression_type=tq, compression_param=dropout, quantization_level=precision))
                        comp_ratios[i] = calculateCompRatio(dataset_array, compress_dataset(dataset_array.copy(), ds_name, andDecompress= False, compression_type= tq, compression_param=dropout, quantization_level=precision))



                np.save('/home/sim/Desktop/TS Extrinsic Regression/data/Quantisation-Level/' + f'qlvl:RMSE-{precision}_{ds_name}_{tq}_{split}', rmses)
                np.save('/home/sim/Desktop/TS Extrinsic Regression/data/Quantisation-Level/' + f'qlvl:CR-{precision}_{ds_name}_{tq}_{split}', comp_ratios)                
```
